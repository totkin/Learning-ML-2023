# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
ARRAY_DIM=10
temp_array=np.zeros(ARRAY_DIM,dtype="int16")
print("init. shape:",np.shape(temp_array),"dim:",temp_array.ndim)
print(temp_array, end="\n\n")

temp_array[4]=1
print("5 element change:")
print("--------| |----------")
print(temp_array, sep="\n", end="\n\n")

SHAPE_TEST=(2,5)
temp_array=temp_array.reshape(SHAPE_TEST)
print("reshape:",np.shape(temp_array),"dim:",temp_array.ndim)
print(temp_array, end="\n\n")
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
temp_array=np.arange(10,50,dtype="int16")[::-1]
print("init:",temp_array, sep="\n", end="\n\n")

print("variant 1: odd position elements",temp_array[1::2], sep="\n", end="\n\n")
print("variant 2: odd elements",temp_array[temp_array%2==0], sep="\n", end="\n\n")
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
temp_array=np.arange(0,9).reshape(3,3)
print("result shape:",np.shape(temp_array),"dim:",temp_array.ndim)
print(temp_array, end="\n\n")
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
temp_array=np.random.rand(4,3,2)
print("init. shape:",np.shape(temp_array),"dim:",temp_array.ndim)
print(temp_array, end="\n\n")

print("max:",temp_array.max(), "min:",temp_array.min())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
temp_array=np.dot(np.random.rand(6,4),np.random.rand(4,3))
print("results")
print("------------------")
print("shape:",np.shape(temp_array),"dim:",temp_array.ndim, end="\n\n")
print(temp_array, end="\n\n")
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
temp_array=np.random.rand(7,7)
print("inti. shape:",np.shape(temp_array),"dim:",temp_array.ndim, end="\n\n")

mean = temp_array.mean()
std_dev = temp_array.std()
min_val = temp_array.min()
max_val = temp_array.max()
norm = np.linalg.norm(temp_array)

print("results")
print("------------------")
print("mean:", mean, sep="    ")
print("average:", np.average(temp_array), sep=" ") # weight=1
print("stddev:", std_dev, sep="  ")
print("min:", min_val, sep="     ")
print("max:", max_val, sep="     ")
print("norm:" ,norm, sep="    ")

print("\n\n")
print("------------------")
print("variant 1: min-max scaling", end="\n\n")
print((temp_array - min_val) / (max_val - min_val))

print("\n\n")
print("------------------")
print("variant 2: z-score", end="\n\n")
print((temp_array - mean) / std_dev)

print("\n\n")
print("------------------")
print("variant 3: L2", end="\n\n")
print(temp_array / norm)
```

### Общие сведения
Данные в источнике полностью заполенны и не имеют пропусков.
В большей части данных, содержатся сведения по мужчинам, не курящим персонам и по событиям классифицированым как "Обед".
Треть случаев приходится на субботы, а всего информация приведена по 4 дням недели с четверга по воскресенье включительно. Наименее наполненный событиями день - пятница.
По распределению размеру чеков и чаевых медианное значение меньше среднего, это указывает на то, что распределение не является симметричным ни в одном, ни в другом случае.

```python
plt.rcParams["figure.figsize"] = [15.00, 5]
plt.rcParams["figure.autolayout"] = True
f, axes = plt.subplots(1, 2)
df = pd.DataFrame(dict(col1=np.linspace(1, 10, 5), col2=np.linspace(1, 10, 5)))

temp_d=pd.pivot_table(tips_raw, index="size", columns="day", values=["total_bill"],aggfunc={"total_bill": "count"})
sns.heatmap(temp_d, annot=True,cmap='coolwarm', ax=axes[0])
temp_d2=pd.pivot_table(tips_raw, index=["size"], columns="day", values=["total_bill"],aggfunc={"total_bill": "mean"})
sns.heatmap(temp_d2, annot=True,cmap='coolwarm', ax=axes[1])

# temp_d=pd.pivot_table(tips_raw, index="day", columns="sex", values=["tip","total_bill","size"],aggfunc={"tip": "sum", 'total_bill': "sum", "size": "count"})
# pd.concat([temp_d, temp_d2], axis=1,join="inner")
```

```python
tips_raw["size_dbl"]=tips_raw["size"]
temp_d3=pd.pivot_table(tips_raw, index=["size"], values=["size_dbl","total_bill","tip","sex"],aggfunc={"size_dbl":"max","total_bill": "sum","tip":"sum","sex":"count"})
temp_d3["tip_percent"]=temp_d3["tip"]/temp_d3["total_bill"]
temp_d3["total_bill_per_capita"]=temp_d3["total_bill"]/temp_d3["sex"]/temp_d3["size_dbl"]
temp_d3["tip_per_capita"]=temp_d3["tip"]/temp_d3["sex"]/temp_d3["size_dbl"]


fig, ax = plt.subplots(figsize=(10,4))
lineplot = sns.lineplot(data=temp_d3, x="size", y="tip_per_capita",label ='tip_per_capita'.replace("_"," "), legend=False)
sns.despine()

ax2 = ax.twinx()
lineplot2 = sns.lineplot(data=temp_d3, x="size", y="total_bill_per_capita", color="tomato",label ='total_bill_per_capita'.replace("_"," "), legend=False) 
sns.despine(right=False)
ax.figure.legend()
```
